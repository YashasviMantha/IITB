```python
input_file = open("Input.csv","r")
input_data = input_file.readlines()
for i in range(len(input_data)):
    input_data[i] = input_data[i].rstrip()
    input_data[i] = input_data[i].split(',')
    input_data[i] = list(map(int,input_data[i]))
input_train = []
input_tune = []
input_test = []
for i in range(len(input_data)):
    if(i<=45):
        input_train.append(input_data[i])
    elif(i<=56 and i>45):
        input_tune.append(input_data[i])
    elif(i<64 and i>56):
        input_test.append(input_data[i])

```

```python
output_file = open("Output.csv","r")
output_data_raw = output_file.readlines()
output_data = []
for i in range(len(output_data_raw)):
    output_data_raw[i] = output_data_raw[i].rstrip()
    output_data_raw[i] = list(map(int,output_data_raw[i]))
for i in output_data_raw:
    for j in i:
        output_data.append(j)
output_train = []
output_tune = []
output_test = []
for i in range(len(input_data)):
    if(i<=45):
        output_train.append(output_data[i])
    elif(i<=56 and i>45):
        output_tune.append(output_data[i])
    elif(i<64 and i>56):
        output_test.append(output_data[i])
```
